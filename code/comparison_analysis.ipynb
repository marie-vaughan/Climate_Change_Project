```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# Load both CSVs
original = pd.read_csv('../ocr_evaluation_initial.csv')
cropped = pd.read_csv('../ocr_evaluation_cropped.csv')

#filter out 'not visible' in columns: material_true, care_true, and country_true
original = original[(original['material_true'] != 'not visible') &
                    (original['care_true'] != 'not visible') &
                    (original['country_true'] != 'not visible')]
cropped = cropped[(cropped['material_true'] != 'not visible') &
                    (cropped['care_true'] != 'not visible') &
                    (cropped['country_true'] != 'not visible')]

# graph counts of country_fuzzy greater than 0

plt.figure(figsize=(10, 6))
sns.histplot(original[original['country_fuzzy'] > 0]['country_fuzzy'], bins=20, color='blue', alpha=0.5, label='Original')
sns.histplot(cropped[cropped['country_fuzzy'] > 0]['country_fuzzy'], bins=20, color='orange', alpha=0.5, label='Cropped')
plt.title('Distribution of Country Fuzzy Match Scores')
plt.xlabel('Fuzzy Match Score')
plt.ylabel('Count')
plt.legend()
plt.show()

#calculate percent of country_fuzzy greater than 0.2
original_above_threshold = (original['country_fuzzy'] > 0).sum() / len(original) * 100
cropped_above_threshold = (cropped['country_fuzzy'] > 0).sum() / len(cropped) * 100
print(f"Original images - % of country_fuzzy > 0: {original_above_threshold:.2f}%")
print(f"Cropped images - % of country_fuzzy > 0: {cropped_above_threshold:.2f}%")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(original[original['material_fuzzy'] > 0]['material_fuzzy'], bins=20, color='blue', alpha=0.5, label='Original')
sns.histplot(cropped[cropped['material_fuzzy'] > 0]['material_fuzzy'], bins=20, color='orange', alpha=0.5, label='Cropped')
plt.title('Distribution of Country Fuzzy Match Scores')
plt.xlabel('Fuzzy Match Score')
plt.ylabel('Count')
plt.legend()
plt.show()

#calculate percent of material_fuzzy
#
# greater than 0.2
original_above_threshold = (original['material_fuzzy'] > 0).sum() / len(original) * 100
cropped_above_threshold = (cropped['material_fuzzy'] > 0).sum() / len(cropped) * 100
print(f"Original images - % of material_fuzzy > 0: {original_above_threshold:.2f}%")
print(f"Cropped images - % of material_fuzzy > 0: {cropped_above_threshold:.2f}%")
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(original[original['care_fuzzy'] > 0]['care_fuzzy'], bins=20, color='blue', alpha=0.5, label='Original')
sns.histplot(cropped[cropped['care_fuzzy'] > 0]['care_fuzzy'], bins=20, color='orange', alpha=0.5, label='Cropped')
plt.title('Distribution of Country Fuzzy Match Scores')
plt.xlabel('Fuzzy Match Score')
plt.ylabel('Count')
plt.legend()
plt.show()

#calculate percent of care_fuzzy
#
# greater than 0.2
original_above_threshold = (original['care_fuzzy'] > 0).sum() / len(original) * 100
cropped_above_threshold = (cropped['care_fuzzy'] > 0).sum() / len(cropped) * 100
print(f"Original images - % of care_fuzzy > 0: {original_above_threshold:.2f}%")
print(f"Cropped images - % of care_fuzzy > 0: {cropped_above_threshold:.2f}%")
```
